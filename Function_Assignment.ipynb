Q - 6  Explain the concept of generators in Python and how they are defined.

Ans. A generator is a function that uses the yield keyword to return values one at a time, pausing between each one, rather than returning all values at once.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

counter = count_up_to(3)

print(next(counter))  # 1
print(next(counter))  # 2
print(next(counter))  # 3
# print(next(counter))  # Raises StopIteration
```

Q - 1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_evens(numbers):
  even_sum = 0
  for num in numbers:
    if num % 2 == 0:
      even_sum += num
  return even_sum

my_list = list(map(int, input("Enter number").split()))
result = sum_of_evens(my_list)
print("sum of even: ", result)
```

Q - 4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime (n):
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n%i==0:
      return False
  return True

num = int(input("Enter the number"))

if 1 <= num <= 200:
  if is_prime(num):
    print(f"{num} is a prime number")
  else:
    print(f"{num} is not a prime number.")
else:
  print("number out of range")
```

```python
is_prime = lambda n: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

prime_num = list(filter(is_prime, range(1, 201)))
print(f"Prime number from 1 to 200 are {prime_num}")
```

Q - 5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

n = int(input("Enter number of Fibonacci terms: "))

fib = Fibonacci(n)

print("Fibonacci sequence:")
for num in fib:
    print(num, end=" ")
```

Q - 6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powerTwo(n):
    for i in range(n):
        yield 2**i

num = int(input("Enter the maximum exponent: "))

print(f"Powers of 2 from 2^0 to 2^{num}:")
for value in powerTwo(num):
    print(value, end=" ")
```

Q - 7  Implement a generator function that reads a file line by line and yields each line as a string.

```python
from io import StringIO

def read_lines(data):
    fake_file = StringIO(data)
    for line in fake_file:
        yield line.rstrip('\n')

data = """Hello
World
This is a simulated file."""

for line in read_lines(data):
    print(line)
```

```python
celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Fahrenheit:", fahrenheit)
```
